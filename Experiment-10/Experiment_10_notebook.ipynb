# **Kohler Illumination Microscopy**

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy as sp
import pandas as pd
```

```python
Intensities_and_Contrasts = {}  # Dictionary to store intensities and contrasts
```

```python
greenlight_data_625lp = np.array(pd.read_csv('greenlight - 625lp_data.csv', encoding='ISO-8859-1'))
whitelight_data_625lp = np.array(pd.read_csv('whitelight - 625lp_data.csv', encoding='ISO-8859-1'))
# print(greenlight_data_625lp)

green_peaks , _ = sp.signal.find_peaks(greenlight_data_625lp[:,1], height=100, distance=20, prominence=50)
green_peaks_distances = greenlight_data_625lp[green_peaks,0]
green_peak_intensities = greenlight_data_625lp[green_peaks,1]
white_peaks , _ = sp.signal.find_peaks(whitelight_data_625lp[:,1], height=200, distance=20, prominence=50)
white_peaks_distances = whitelight_data_625lp[white_peaks,0]
white_peak_intensities = whitelight_data_625lp[white_peaks,1]

green_valleys , _ = sp.signal.find_peaks(-greenlight_data_625lp[:,1], height=-100, distance=20, prominence=50)
green_valleys_distances = greenlight_data_625lp[green_valleys,0]
green_valley_intensities = greenlight_data_625lp[green_valleys,1]
white_valleys , _ = sp.signal.find_peaks(-whitelight_data_625lp[:,1], height=-100, distance=20, prominence=50)
white_valleys_distances = whitelight_data_625lp[white_valleys,0]
white_valley_intensities = whitelight_data_625lp[white_valleys,1]


fig, axs = plt.subplots(2, 1, figsize=(12, 12))
axs[0].plot(greenlight_data_625lp[:,0], greenlight_data_625lp[:,1], label='Green Light with 6.25 lp/mm Filter', color='green')
axs[0].plot(green_peaks_distances, green_peak_intensities, "o", label='Peaks', color='red')
axs[0].plot(green_valleys_distances, green_valley_intensities, "o", label='Valleys', color='blue')
axs[0].set_title('Intensity vs Distance for Green Light with 6.25 lp/mm Filter')
axs[0].set_xlabel('Distance (um)')
axs[0].set_ylabel('Intensity (a.u.)')
axs[0].legend()
axs[0].grid()
axs[1].plot(whitelight_data_625lp[:,0], whitelight_data_625lp[:,1], label='White Light with 6.25 lp/mm Filter', color='orange')
axs[1].plot(white_peaks_distances, white_peak_intensities, "o", label='Peaks', color='red')
axs[1].plot(white_valleys_distances, white_valley_intensities, "o", label='Valleys', color='blue')
axs[1].set_title('Intensity vs Distance for White Light with 6.25 lp/mm Filter')
axs[1].set_xlabel('Distance (um)')
axs[1].set_ylabel('Intensity (a.u.)')
axs[1].legend()
axs[1].grid()
plt.show()
```

```python
I_max_green = np.max(greenlight_data_625lp[:,1])
I_min_green = np.min(greenlight_data_625lp[:,1])
contrast_green = (I_max_green - I_min_green) / (I_max_green + I_min_green)
print(f'Contrast for Green Light with 6.25 lp/mm Filter: {contrast_green:.4f}')
I_max_white = np.max(whitelight_data_625lp[:,1])
I_min_white = np.min(whitelight_data_625lp[:,1])    
contrast_white = (I_max_white - I_min_white) / (I_max_white + I_min_white)
print(f'Contrast for White Light with 6.25 lp/mm Filter: {contrast_white:.4f}')

Intensities_and_Contrasts['6.25 lp/mm'] = {
    'Green': {
        'I_max': I_max_green,
        'I_min': I_min_green,
        'Contrast': contrast_green
    },
    'White': {
        'I_max': I_max_white,
        'I_min': I_min_white,
        'Contrast': contrast_white
    }
}
```
