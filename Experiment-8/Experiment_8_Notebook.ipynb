# Calculations

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy as sp
```

### Absorption Spectrum

```python
absorption_data = pd.read_excel('absorption_spectrum.xlsx')
absorption_data_values = np.array(absorption_data[5:])

Wavelengths = absorption_data_values[:,0]
Absorbance = absorption_data_values[:,1]

plt.figure(figsize=(12,8))
plt.plot(Wavelengths, Absorbance, label='Raw Data')

Wavelength_mask = (Wavelengths > 500) & (Wavelengths < 550)
Wavelengths = Wavelengths[Wavelength_mask]
Absorbance = Absorbance[Wavelength_mask]


plt.plot(Wavelengths, Absorbance, label='Used Data', color='orange')
plt.xlabel('Wavelength (nm)')
plt.ylabel('Absorbance')
plt.title('Absorption Spectrum')
plt.legend()
plt.show()
```

```python
# Peak finding using prominence and distance
peaks, properties = sp.signal.find_peaks(
    Absorbance,
    prominence=0.1,      # adjust if needed
    distance=5           # enforces one peak per band
)

# Extract wavelengths and absorbance of filtered peaks
peak_wavelengths = Wavelengths[peaks]
peak_absorbance = Absorbance[peaks]

# print("Selected band-head wavelengths (nm):")
# print(peak_wavelengths)

plt.figure(figsize=(12,8))
plt.plot(Wavelengths, Absorbance, label='Spectrum')
plt.plot(peak_wavelengths, peak_absorbance, 'rx', label='Band heads')
plt.xlabel('Wavelength (nm)')
plt.ylabel('Absorbance')
plt.title('Iodine Absorption Spectrum with Band Heads')
plt.legend()
plt.show()
```

```python
wavenumbers = 1 / (peak_wavelengths * 1e-7)

# Locate index of band head near 543 nm
idx_543 = np.where((peak_wavelengths > 542.5) & (peak_wavelengths < 543.5))[0]

if len(idx_543) == 0:
    raise ValueError("No peak found near 543 nm")

idx_543 = idx_543[0]

# print("Index of 543 nm band head:", idx_543)
# print("Wavelength at index:", peak_wavelengths[idx_543])
# print("Wavenumber at index:", wavenumbers[idx_543])

v_at_543 = 26

# Total number of selected band heads
N = len(wavenumbers)

# Create vibrational quantum number array
v_numbers = np.zeros(N, dtype=int)

# Assign v' = 26 at the 543 nm index
v_numbers[idx_543] = v_at_543

# Fill higher-energy side (lower index ‚Üí higher v')
for i in range(idx_543 - 1, -1, -1):
    v_numbers[i] = v_numbers[i + 1] + 1

# Fill lower-energy side (higher index ‚Üí lower v')
for i in range(idx_543 + 1, N):
    v_numbers[i] = v_numbers[i - 1] - 1

# for wl, v in zip(peak_wavelengths, v_numbers):
#     print(f"{wl:.1f} nm  ‚Üí  v' = {v}")
```

```python
vibrational_levels = v_numbers
# print("Vibrational levels: ", vibrational_levels)
# print(len(vibrational_levels), len(wavenumbers))

wavenumbers = wavenumbers[::-1]
vibrational_levels = vibrational_levels[::-1]

# del_G = wavenumbers[1:] - wavenumbers
del_G = np.diff(wavenumbers)
del_2_G = np.diff(del_G)
# print("Delta G (cm^-1): ", del_G)

v_levels = np.asarray(vibrational_levels[:-1], dtype=float)
del_G = np.asarray(del_G, dtype=float)
# Linear fit
slope, intercept = np.polyfit(v_levels, del_G, 1)

v_levels_2 = np.asarray(vibrational_levels[:-2], dtype=float)
del_2_G = np.asarray(del_2_G, dtype=float)
# Second derivative fit (optional)
slope_2, intercept_2 = np.polyfit(v_levels_2, del_2_G, 1)

# Birge‚ÄìSponer plot
plt.figure(figsize=(8,6))
plt.scatter(v_levels, del_G)
plt.plot(v_levels, slope * v_levels + intercept, 'r-', label='Linear fit')
plt.xlabel(r'$v$')
plt.ylabel(r'$\Delta G$ (cm$^{-1}$)')
plt.title('Birge Sponer Plot (Iodine)')
plt.grid(True)
plt.legend()
plt.show()

plt.figure(figsize=(8,6))
plt.scatter(v_levels_2, del_2_G)
plt.plot(v_levels_2, slope_2 * v_levels_2 + intercept_2, 'r-', label='Linear fit')
plt.xlabel(r'$v$')
plt.ylabel(r'$\Delta^2 G$ (cm$^{-1}$)')
plt.title('Second Derivative Birge Sponer Plot (Iodine)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Vibrational constants
omega_0 = intercept - 0.5 * slope
xi_0 = -slope / (2 * omega_0)
D_0 = omega_0**2 / (4 * xi_0 * omega_0)

print(f"Slope = {slope:.3f}")
print(f"Intercept = {intercept:.2f} cm^-1")
print(f"Fundamental frequency of vibration, œâ_0 = {omega_0:.2f} cm^-1")
print(f"Anharmonicity constant, x_0 = {xi_0:.4f} cm^-1")
print(f"Dissociation energy, D_0 = {D_0:.2f} cm^-1")
```

### Fluorescence Spectrum

```python
with open('Flourscence_group16.txt', 'r') as f:
    flourescence_data = f.readlines()
flourescence_wavelengths = []
flourescence_intensities = []   
for line in flourescence_data:
    wl, inten = line.strip().split()
    flourescence_wavelengths.append(float(wl))
    flourescence_intensities.append(float(inten))


flourescence_wavelengths = np.array(flourescence_wavelengths[55:])
flourescence_intensities = np.array(flourescence_intensities[55:])
plt.figure(figsize=(12,8))
plt.plot(flourescence_wavelengths, flourescence_intensities, label='Fluorescence Spectrum', color='green')
plt.xlabel('Wavelength (nm)')
plt.ylabel('Intensity (a.u.)')
plt.title('Iodine Fluorescence Spectrum')
plt.legend()
plt.show()
```

```python
# Peak finding using prominence and distance
peaks, properties = sp.signal.find_peaks(
    flourescence_intensities,
    prominence=250,      # adjust if needed
    distance=5           # enforces one peak per band
)   

# Extract wavelengths and absorbance of filtered peaks
peak_wavelengths = flourescence_wavelengths[peaks]
peak_intensities = flourescence_intensities[peaks]

# print("Selected band-head wavelengths (nm):")
# print(peak_wavelengths)

plt.figure(figsize=(12,8))
plt.plot(flourescence_wavelengths, flourescence_intensities, label='Spectrum')
plt.plot(peak_wavelengths, peak_intensities, 'rx', label='Peaks')
plt.xlabel('Wavelength (nm)')
plt.ylabel('Intensity (a.u.)')
plt.title('Iodine Fluorescence Spectrum with Peaks')
plt.legend()
plt.show()
```

```python
# print("Peak Wavelengths (nm):")
# print(peak_wavelengths)

E_wavelengths = 1 / (peak_wavelengths * 1e-7)
E_532 = 1 / (532 * 1e-7)
E_diff = E_532 - E_wavelengths
Vibrational_numbers = np.arange(1,len(E_diff)+1)
# Vibrational_numbers = Vibrational_numbers - 0.5

print(len(Vibrational_numbers), len(E_diff))

# print("Vibrational Numbers:")
# print(Vibrational_numbers)
# print("Energy Differences (cm^-1):")
# print(E_diff)

coefficients = np.polyfit(Vibrational_numbers, E_diff, 3)

plt.figure(figsize=(12,8))
plt.scatter(Vibrational_numbers, E_diff, label='Data Points')
plt.plot(Vibrational_numbers, np.polyval(coefficients, Vibrational_numbers), 'r-', label='Linear Fit')
plt.xlabel('Vibrational Quantum Number v\'\'')
plt.ylabel('Energy Difference (cm$^{-1}$)')
plt.title('Vibrational Energy Levels from Fluorescence Data')
plt.legend()
plt.show()

# print(coefficients)
```

```python
print("Coefficients of Dunham expansion (from fluorescence data):")
print(f"Y00: {coefficients[3]:.4f} cm^-1")
print(f"Y10: {coefficients[2]:.4f} cm^-1")
print(f"Y20: {coefficients[1]:.4f} cm^-1")
print(f"Y30: {coefficients[0]:.6f} cm^-1")
```
